```python
import cvxpy as cp
import numpy as np


def optimize_portfolio(returns, lambda_reg, lasso=False):
    # Compute the covariance matrix
    cov_matrix = returns.cov()
    n_assets = returns.shape[1]
    
    # Define the weights as a CVXPY variable
    w = cp.Variable(n_assets)
    
    # Portfolio variance
    portfolio_variance = cp.quad_form(w, cov_matrix.values)
    
    # Define the regularization term
    regularization = cp.norm(w, 1) if lasso else cp.sum_squares(w)
    
    # Objective function with regularization
    objective = cp.Minimize(portfolio_variance + lambda_reg * regularization)
    
    # Constraints: sum of weights = 1
    constraints = [cp.sum(w) == 1]
    
    # Define and solve the problem
    problem = cp.Problem(objective, constraints)
    problem.solve()
    
    print("Problem status:", problem.status)
    
    # Check if the solution is optimal
    if problem.status != cp.OPTIMAL:
        print("The optimization problem did not solve optimally. Status:", problem.status)
        return None
    
    return w.value



# Assuming you have a DataFrame called 'returns' with your assets' returns
# Here, 'returns' would be the DataFrame that you have cleaned and prepared earlier

# Define a range of lambda values you want to test, for example:
lambda_values = np.arange(0.01, 0.1, 0.01)

# Initialize dictionaries to store the optimal weights for each lambda value
optimal_weights_ridge = {}
optimal_weights_lasso = {}

# Loop over lambda values to find the optimal portfolio weights
for lambda_reg in lambda_values:
    optimal_weights_ridge[lambda_reg] = optimize_portfolio(returns, lambda_reg, lasso=False)
    optimal_weights_lasso[lambda_reg] = optimize_portfolio(returns, lambda_reg, lasso=True)

# Now you can analyze the weights, variances, etc. for different values of lambda

```

```python
import cvxpy as cp
import numpy as np

def optimize_portfolio_without_regularization(returns):
    # Compute the covariance matrix
    cov_matrix = returns.cov()
    n_assets = returns.shape[1]
    
    # Define the weights as a CVXPY variable
    w = cp.Variable(n_assets)
    
    # Portfolio variance
    portfolio_variance = cp.quad_form(w, cov_matrix.values)
    
    # Objective function is simply to minimize the portfolio variance
    objective = cp.Minimize(portfolio_variance)
    
    # Constraints: sum of weights = 1
    constraints = [cp.sum(w) == 1]
    
    # Define and solve the problem
    problem = cp.Problem(objective, constraints)
    
    # Solve the problem using a solver that supports quadratic programming
    problem.solve()
    
    print("Problem status:", problem.status)
    
    # Check if the solution is optimal
    if problem.status in [cp.OPTIMAL, cp.OPTIMAL_INACCURATE]:
        return w.value
    else:
        print("The optimization problem did not solve optimally. Status:", problem.status)
        return None

# Assuming 'returns' is your DataFrame containing return data for the assets
optimal_weights_no_regularization = optimize_portfolio_without_regularization(returns)

if optimal_weights_no_regularization is not None:
    print("Optimization successful. Weights obtained without regularization.")
    # Print the obtained optimal weights
    print(optimal_weights_no_regularization)
else:
    print("Failed to optimize portfolio without regularization.")
```

```python
def optimize_portfolio_no_short_selling(returns):
    cov_matrix = returns.cov()
    n_assets = returns.shape[1]
    
    w = cp.Variable(n_assets)
    portfolio_variance = cp.quad_form(w, cov_matrix.values)
    
    # Adding a non-negativity constraint for no short selling
    constraints = [cp.sum(w) == 1, w >= 0]
    
    problem = cp.Problem(cp.Minimize(portfolio_variance), constraints)
    problem.solve()
    
    if problem.status in [cp.OPTIMAL, cp.OPTIMAL_INACCURATE]:
        return w.value
    else:
        print("Optimization with no short selling did not solve optimally. Status:", problem.status)
        return None

optimal_weights_no_short = optimize_portfolio_no_short_selling(returns)

if optimal_weights_no_short is not None:
    print("Optimal weights without short selling:", optimal_weights_no_short)
else:
    print("Failed to obtain optimal weights without short selling.")
```

```python
import numpy as np
from sklearn.linear_model import Ridge, Lasso
from sklearn.model_selection import cross_val_score, KFold
from sklearn.datasets import make_regression

# Simulate some regression data
X, y = make_regression(n_samples=100, n_features=20, noise=0.1, random_state=42)

# Define a range of lambda (alpha) values
lambda_values = np.logspace(-4, 4, 100)

# Setup cross-validation
kf = KFold(n_splits=5, shuffle=True, random_state=42)

# Initialize arrays to store results
ridge_cv_results = []
lasso_cv_results = []

# Perform CV for Ridge
for alpha in lambda_values:
    ridge = Ridge(alpha=alpha)
    scores = cross_val_score(ridge, X, y, cv=kf, scoring='neg_mean_squared_error')
    ridge_cv_results.append(scores.mean())

# Perform CV for Lasso
for alpha in lambda_values:
    lasso = Lasso(alpha=alpha)
    scores = cross_val_score(lasso, X, y, cv=kf, scoring='neg_mean_squared_error')
    lasso_cv_results.append(scores.mean())

# Convert to positive mean squared error
ridge_cv_results = np.array(ridge_cv_results) * -1
lasso_cv_results = np.array(lasso_cv_results) * -1

# Now you can plot the results or find the optimal lambda value
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(lambda_values, ridge_cv_results, label='Ridge CV Results')
plt.plot(lambda_values, lasso_cv_results, label='Lasso CV Results')
plt.xscale('log')  # Lambda values are on a log scale
plt.xlabel('Lambda')
plt.ylabel('Mean Squared Error')
plt.title('Hyperparameter Tuning via Cross-Validation')
plt.legend()
plt.show()
```
