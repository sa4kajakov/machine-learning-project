```python
import sys
!{sys.executable} -m pip install pandas
import pandas as pd 

import sys
!{sys.executable} -m pip install matplotlib seaborn
import matplotlib.pyplot as plt


import sys
!{sys.executable} -m pip install numpy
import numpy as np


from os import path

import sys
!{sys.executable} -m pip install -U scikit-learn

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import precision_score, recall_score, f1_score, accuracy_score, classification_report, confusion_matrix
from sklearn.metrics import roc_curve, auc

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
display(data)
```

```python
#Data Cleaning
data_clean = data.dropna()
display(data_clean)
```

```python
statistics = data_clean.describe().round(1)
display(statistics)
```

```python
# 70% training, 30% testing
X = data_clean.drop('Default Indicator', axis=1)
y = data_clean['Default Indicator']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
# Training the logistic regression model
model = LogisticRegression(max_iter=2000)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
precision = precision_score(y_test, y_pred, average='macro')
print("Precision:", precision)
```

```python
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
recall = recall_score(y_test, y_pred)
print("Recall:", recall)
```

```python
f1 = f1_score(y_test, y_pred)
print("F1 Score:", f1)
```

```python
cr = classification_report(y_test, y_pred)
print(cr)
```

```python
conf_matrix = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt="d", cmap='Blues', 
            xticklabels=['0', '1'], 
            yticklabels=['0', '1'])
plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.title('Confusion Matrix')
plt.show()
```

```python
sns.set_theme(style="whitegrid")

plt.figure(figsize=(8, 6))
ax = sns.countplot(x='Default Indicator', data=data_clean)

for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', fontsize=11, color='gray', xytext=(0, 5),
                textcoords='offset points')

plt.title('Banks debt Distribution', fontsize=15)
plt.ylabel('Count', fontsize=12)
plt.xlabel('')
plt.xticks([0, 1], ['Loans not defaulted - 0', 'Loans defaulted - 1'], fontsize=10)
plt.show()
```

```python
from sklearn.utils import resample

# Separate the classes
df_majority = data_clean[data_clean['Default Indicator'] == 0]
df_minority = data_clean[data_clean['Default Indicator'] == 1]

# Downsample majority class
df_majority_downsampled = resample(df_majority, 
                                   replace=False,    # sample without replacement
                                   n_samples=len(df_minority),     # to match minority class size
                                   random_state=123) # reproducible results

# Combine minority class with downsampled majority class
df_balanced = pd.concat([df_majority_downsampled, df_minority])

# Display new class counts
print(df_balanced['Default Indicator'].value_counts())

# Visualize the new class distribution
sns.set_theme(style="whitegrid")
plt.figure(figsize=(8, 6))
ax = sns.countplot(x='Default Indicator', data=df_balanced)

for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', fontsize=11, color='gray', xytext=(0, 5),
                textcoords='offset points')

plt.title('Balanced Banks debt Distribution After Undersampling', fontsize=15)
plt.ylabel('Count', fontsize=12)
plt.xlabel('')
plt.xticks([0, 1], ['Loans not defaulted - 0', 'Loans defaulted - 1'], fontsize=10)
plt.show()



```

```python
test_combined = X_test.copy()
test_combined['Default Indicator'] = y_test

```

```python
df_majority_test = test_combined[test_combined['Default Indicator'] == 0]
df_minority_test = test_combined[test_combined['Default Indicator'] == 1]

df_majority_test_downsampled = resample(df_majority_test, 
                                         replace=False,    
                                         n_samples=len(df_minority_test),     
                                         random_state=123) 

df_test_balanced = pd.concat([df_majority_test_downsampled, df_minority_test])
```

```python
X_test_balanced = df_test_balanced.drop('Default Indicator', axis=1)
y_test_balanced = df_test_balanced['Default Indicator']
```

```python
y_pred_balanced = model.predict(X_test_balanced)
```

```python
accuracy = accuracy_score(y_test_balanced, y_pred_balanced)
print("Accuracy:", accuracy)
precision = precision_score(y_test_balanced, y_pred_balanced, average='macro')
print("Precision:", precision)
recall = recall_score(y_test_balanced, y_pred_balanced)
print("Recall:", recall)
f1 = f1_score(y_test_balanced, y_pred_balanced)
print("F1 Score:", f1)
```

```python
conf_matrix = confusion_matrix(y_test_balanced, y_pred_balanced)

plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt="d", cmap='Blues', 
            xticklabels=['0', '1'], 
            yticklabels=['0', '1'])
plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.title('Confusion Matrix')
plt.show()
```

```python
y_pred_prob = model.predict_proba(X_test_balanced)[:, 1]  # Probabilities for the positive class
fpr, tpr, thresholds = roc_curve(y_test_balanced, y_pred_prob)
roc_auc = auc(fpr, tpr)
```

```python
# Plotting the ROC Curve
plt.figure()
plt.plot(fpr, tpr, color='red', lw=2, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='blue', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend(loc="lower right")
plt.show()
```
